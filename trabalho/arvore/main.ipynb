# Classificação de Cogumelos com Árvore de Decisão

**Objetivo:** Prever se um cogumelo é **venenoso (p)** ou **comestível (e)** com base em suas características físicas, utilizando o algoritmo de Árvore de Decisão.

**Dataset:** [Mushroom Dataset (UCI)](https://www.kaggle.com/datasets/uciml/mushroom-classification).

---

### O que é uma Árvore de Decisão?

A **Árvore de Decisão** aprende uma série de **perguntas encadeadas** sobre os dados para chegar a uma classificação.

## Vantagens em relação ao KNN:

* Não precisa de normalização dos dados
* Resultado interpretável, você consegue visualizar as decisões
* Lida bem com variáveis categóricas e numéricas juntas

## Importação das bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

---
## 1. Carregamento dos Dados

```python
df = pd.read_csv('../data.csv')

print(f'Colunas lidas: {list(df.columns)}\n')
df
```

---
## 2. Limpeza e Preparação dos Dados

```python

FEATURES = [ col for col in df.columns if col != 'class' ]
TARGET = 'class'

le_features = {}
for col in FEATURES:
    le_col = LabelEncoder()
    df[col] = le_col.fit_transform(df[col].astype(str))
    le_features[col] = le_col

print(f'Serão no total {len(FEATURES)} features utilizadas')
```

### 2.1 Distribuição do Target

```python
df[TARGET] = df[TARGET].astype(str).str.strip().str.lower()
 
df = df.dropna(subset=FEATURES + [TARGET]).copy()

display(df.head())

print('Distribuição do target:')
print(df[TARGET].value_counts())
```

---
## 3. Escolha e Preparação das Features

```python
le = LabelEncoder()

y_all = le.fit_transform(df[TARGET])
X_all = df[FEATURES].values

print(f'Classes: {list(le.classes_)}')
print(f'Features: {FEATURES}')
```

---
## 4. Separação em Treino e Teste

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_all, y_all,
    test_size=0.2,
    random_state=42,
    stratify=y_all
)

print(f'Treino: {len(X_train)/len(X_all)*100:.0f}% = {len(X_train)} amostras')
print(f'Teste: {len(X_test)/len(X_all)*100:.0f}% = {len(X_test)} amostras')
```

---
## 5. Treinamento com a Árvore de Decisão

```python
model = DecisionTreeClassifier(
    max_depth=5, random_state=42
)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

---
## 6. Avaliação do Modelo

```python
print(f"Acurácia: {accuracy_score(y_test, y_pred):.2%}")
print("\nRelatório de classificação:")
print(classification_report(y_test, y_pred, target_names=le.classes_))

print("Matriz de confusão:")
print(confusion_matrix(y_test, y_pred))

plt.figure(figsize=(20, 8))
plot_tree(
    model,
    feature_names=FEATURES,
    class_names=list(le.classes_),
    filled=True,
    rounded=True,
    fontsize=9,
)

plt.title('Árvore de Decisão', fontsize=16, fontweight='bold', pad=20)
plt.show()
```

```python
print('Exemplos de Previsões:')
sample = pd.DataFrame({
    'Real'     : le.inverse_transform(y_test),
    'Previsto' : le.inverse_transform(y_pred),
}).reset_index(drop=True)
print(sample.head(10).to_string(index=False))
```
