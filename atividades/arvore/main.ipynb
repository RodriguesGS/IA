# Classificação de Gripe com Árvore de Decisão

**Objetivo**: Prever se uma pessoa ficou gripada no ano passado com base em seus hábitos de saúde e estilo de vida, utilizando o algoritmo de Árvore de Decisão.

O que é uma Árvore de Decisão?
A Árvore de Decisão é um algoritmo de aprendizado supervisionado que aprende uma série de perguntas encadeadas sobre os dados para chegar a uma classificação.

## Vantagens em relação ao KNN:

* Não precisa de normalização dos dados
* Resultado interpretável, você consegue visualizar as decisões
* Lida bem com variáveis categóricas e numéricas juntas
